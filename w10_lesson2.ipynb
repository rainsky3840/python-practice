```python
from matplotlib import pyplot as plt
```

```python
import random
x=[i for i in range(1,101)]
y=[]
for i in range(100):
  y.append(random.randint(1,101))
```

```python
plt.plot(x,y)
plt.show() #ipynb notebook에서 실행하지 않을 경우
```

```python
plt.scatter(x,y)
plt.show() #ipynb notebook에서 실행하지 않을 경우
```

```python
# Colors: w r y g k c b m
# Shapes: o p s d h x v . , 1 2 3 4 8 - = | : < > -- -.
plt.plot(x,y, 'rh')
plt.title('Some title')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

```python
lines = plt.plot(x,y)
plt.setp(lines, linewidth=5, color='g')
plt.show()
```

```python
# Multiple Graphs
y_1 = []
y_2 = []
for i in range(100):
  y_1.append(random.randint(1,101))
  y_2.append(random.randint(1,101))
```

```python
plt.plot(x, y_1, 'gx', x, y_2, 'ro')
plt.show()
```

```python
plt.plot(x, y_1, 'g', x, y_2, 'r')
plt.show()
```

```python
#plt.subplot(row, col, index)
plt.subplot(2, 1, 1)
plt.title('Graph 1')
plt.plot(x,y_1,'r')
plt.subplot(2, 1, 2)
plt.title('Graph 2')
plt.plot(x,y_2,'g')
plt.suptitle('Graphs') #super title
plt.show()
```

```python
#plt.subplot(row, col, index)
plt.subplot(1, 2, 1)
plt.title('Graph 1')
plt.plot(x,y_1,'r')
plt.subplot(1, 2, 2)
plt.title('Graph 2')
plt.plot(x,y_2,'g')
plt.suptitle('Graphs') #super title
plt.show()
```

```python
bar_chart_data = {'one': 5, 'two': 10, 'three': 7}
plots = plt.bar(bar_chart_data.keys(), bar_chart_data.values())
plt.setp(plots, color='m')
plt.title('Bar Graph')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
#7) Pie Chart
pie_chart_data = {'one':30, 'two':20, 'three':10} #each numbers are portion of whole
explode = (0, 0, 0.2)
plt.pie(pie_chart_data.values(), labels=pie_chart_data.keys(), startangle=90, explode=explode, shadow=True)
```

```python
histogram_ds = []
for i in range(20):
  histogram_ds.append(random.randint(1,20))
print(histogram_ds)

plt.title('Histogram')
plt.xlabel('Count')
plt.ylabel('Numbers 1-20')
plt.hist(histogram_ds, color='g', cumulative=True, density=True)
#cumulative: 누적 / density: proportion that each category is taking up
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D #plot on 3d set of axes
axes = plt.axes(projection='3d')
axes.scatter3D(x,y_1, y_2, marker='^') #색깔은 못 바꾸지만 모양은 가능
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_zlabel('Z axis')
plt.show()
```
