```python
from matplotlib import pyplot as plt
import random
```

```python
x = [i for i in range(1,101)]
y_1 = []
y_2 = []

for i in range(100):
  y_1.append(random.randint(1,101))
  y_2.append(random.randint(1,101))
```

```python
plt.subplot(2,1,1)
plt.title('graph 1')
plt.plot(x, y_1, 'r')

plt.subplot(2,1,2)
plt.title('graph 2')
plt.plot(x, y_2, 'g')

plt.suptitle('Graphs')

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,8))

ax1.plot(x, y_1, 'r')
ax1.set_title('Graph 1')

ax2.plot(x, y_2, 'g')
ax1.set_title('Graph 2')
```
